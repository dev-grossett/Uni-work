```python
import numpy as np
import matplotlib.pyplot as plt

def dF(r, theta):
    return r*(r-4), 1 + 0*theta

X, Y = np.meshgrid(np.linspace(-6.0, 6.0, 40), np.linspace(-6.0, 6.0, 40))
R, Theta = (X**2 + Y**2)**0.5, np.arctan2(Y, X)
dR, dTheta = dF(R, Theta)
C, S = np.cos(Theta), np.sin(Theta)
U, V = dR*C - R*S*dTheta, dR*S+R*C*dTheta

plt.streamplot(X, Y, U, V, color='r', linewidth=0.5, density=1.2)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.show()
```
